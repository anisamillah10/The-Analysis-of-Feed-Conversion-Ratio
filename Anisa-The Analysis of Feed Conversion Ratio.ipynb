```python
duplicates = data.duplicated()
data_duplicates = data[duplicates]
```

```python
data_duplicates
```

```python
# Memisahkan baris duplikat dan non-duplikat
data_duplicates = data[duplicates]
data_non_duplicates = data[~duplicates]

# Menyimpan satu baris duplikat dan menghapus dua lainnya
data_duplicates_to_keep = data_duplicates.drop_duplicates(keep='first')

# Gabungkan kembali DataFrame
data_no_dupes = pd.concat([data_non_duplicates, data_duplicates_to_keep]).reset_index(drop=True)

# Menghapus baris dengan nilai null
data_final = data_no_dupes.dropna()
```

```python
import numpy as np
```

```python
# Menambahkan kolom segmentasi FCR dengan nilai default 'undefined'
data_final = data_final.assign(FCR_segment='undefined')

# Menggunakan .loc untuk menetapkan nilai sesuai kondisi: maksimal FCR ideal adalah 1,5
data_final.loc[data_final['fcr'] < 1.5, 'FCR_segment'] = 'low'
data_final.loc[data_final['fcr'] >= 1.5, 'FCR_segment'] = 'high'

# Menampilkan DataFrame dengan kolom segmentasi
data_final.head(3)
```

```python
data_final['status'].value_counts()
```

```python
# Mengelompokkan data berdasarkan Status dan FCR, kemudian hitung jumlah unik siklus
grouped_data = data_final.groupby(['status', 'FCR_segment']).agg({'cycle_id':'nunique'}).reset_index()

# Menghitung jumlah siklus berdasarkan Status
grouped_data2 = data_final.groupby(['status'])['cycle_id'].count().reset_index()

# Menggabungkan data unik tiap kelompok siklus berdasarkan Status
merge = grouped_data.merge(grouped_data2, on='status')

# Menghitung persentase dari jumlah unik siklus berdasarkan Status dan Churn
merge['Percentage'] = round(merge['cycle_id_x']/merge['cycle_id_y']*100,3)

merge
```

```python
# Menghapus baris yang memiliki nilai 'failed' pada kolom 'status' menggunakan query
merge = merge.query('status != "Failed"')
```

```python
merge
```

```python
# Menghapus kolom 'pond_type'
data_final = data_final.drop(columns=['pond_type'])

# Membuat kolom 'pond_type' baru berdasarkan kriteria yang disebutkan
conditions = [
    (data_final['stocking_density'] > 200) & (data_final['area'].between(0, 2000)),
    (data_final['stocking_density'] <= 200) & (data_final['area'].between(2000, 5000)),
    (data_final['stocking_density'] <= 100) & (data_final['area'].between(5000, 30000)),
    (data_final['stocking_density'] <= 50) & (data_final['area'].between(30000, 100000))
]
choices = ['supra intensif', 'intensif', 'semi intensif', 'ekstentif']

data_final['pond_type'] = np.select(conditions, choices, default='ekstentif')#deafault ekstentif karena termasuk budidaya yang tidka teratur

# Menampilkan DataFrame setelah perubahan
print("DataFrame setelah perubahan:")
data_final.head(3)
```

```python
categories_counts = data_final['pond_type'].value_counts()
categories_counts
```

```python
# Mengelompokkan data berdasarkan Pond Type dan FCR, kemudian hitung jumlah unik siklus
grouped_data3 = data_final.groupby(['pond_type', 'FCR_segment']).agg({'cycle_id':'nunique'}).reset_index()

# Menghitung jumlah siklus berdasarkan Status
grouped_data4 = data_final.groupby(['pond_type'])['cycle_id'].count().reset_index()

# Menggabungkan data unik tiap kelompok siklus berdasarkan Status
merge2 = grouped_data3.merge(grouped_data4, on='pond_type')

# Menghitung persentase dari jumlah unik siklus berdasarkan Status dan Churn
merge2['Percentage'] = round(merge2['cycle_id_x']/merge2['cycle_id_y']*100,3)

merge2
```

```python
# Menentukan ukuran gambar 
plt.figure(figsize=(10,6))

# Membuat bar plot
sns.barplot(x='pond_type', y='Percentage', hue='FCR_segment', data=merge2, palette=['gray', 'darkblue'])

# Menambahkan label pada sumbu x dan y 
plt.xlabel('Pond Type',fontsize=15, fontweight='bold')
plt.ylabel('Percentage',fontsize=15, fontweight='bold')

# Memberikan label pada grafik batang
plt.text(x=-0.20, y=59.4, s='57.988', ha='center', va='center', fontsize=11)
plt.text(x=0.20, y=42.3, s='40.828', ha='center', va='center', fontsize=11)
plt.text(x=0.80, y=65.5, s='64.088', ha='center', va='center', fontsize=11)
plt.text(x=1.20, y=37.4, s='35.912', ha='center', va='center', fontsize=11)
plt.text(x=1.80, y=48.7, s='47.222', ha='center', va='center', fontsize=11)
plt.text(x=2.20, y=54.2, s='52.778', ha='center', va='center', fontsize=11)

# Memberikan judul pada grafik
plt.title('Percentage of Unique Cycle by Pond Type and FCR Segment',fontsize=18, fontweight='bold')
 
plt.show()
```

```python
# Load data dari file csv ke dalam DataFrame pandas
measurements = pd.read_csv('measurements.csv')
```

```python
# Menggabungkan kedua DataFrame secara horizontal hanya dengan kolom 'FCR'
measurements = pd.merge(measurements, data_final[['cycle_id', 'fcr']], on='cycle_id')
```

```python
measurements.info()
```

```python
measurements.isna().sum()
```

```python
# Menentukan threshold jumlah baris NaN
threshold = 10000

# Menghitung jumlah NaN di setiap kolom
na_counts = measurements.isna().sum()

# Menentukan kolom yang memiliki lebih dari 10.000 NaN
cols_to_drop = na_counts[na_counts > threshold].index

# Menghapus kolom-kolom tersebut
measurements2 = measurements.drop(columns=cols_to_drop)
```

```python
measurements2.head(3)
```

```python
# Memisahkan kolom numerik dan non-numerik
numeric_cols2 = measurements2.select_dtypes(include=[np.number]).columns
non_numeric_cols = measurements2.select_dtypes(exclude=[np.number]).columns

# Menghitung median hanya untuk kolom-kolom numerik
medians = measurements2[numeric_cols2].median()

# Mengisi nilai NaN pada kolom numerik dengan median
measurements2[numeric_cols2] = measurements2[numeric_cols2].fillna(medians)
```

```python
measurements2.info()
```

```python
# Melihat korelasi antar kolom menggunakan .corr()
numeric_cols2 = measurements2.select_dtypes(include=[np.number])
correlation_matrix2 = numeric_cols2.corr()
```

```python
# Buat heatmap korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix2, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap Korelasi Antar Kolom')
plt.show()
```

**Interpretasi**<br>
Grafik di atas menunjukkan korelasi tiap tabel numerik dengan tabel lainnya. Kolom yang memiliki korelasi positif adalah kolom morning_salinity dan morning_temperature, morning_salinity dan morning_ph. Artinya semakin tinggi morning_salinity maka semakin tinggi morning_temperature dan semakin tinggi morning_salinity maka semakin tinggi morning_ph. 

```python
# Menghitung korelasi antara 'fcr' dengan kolom lainnya
correlations = numeric_cols2.corr()['fcr'].drop('fcr')

# Membuat bar plot
correlations.plot(kind='bar')
plt.title('Korelasi antara FCR dan Parameter kualitas air')
plt.ylabel('Korelasi Pearson')
plt.show()

# Menyimpan plot sebagai gambar
plt.savefig('grafik_sin.png')
```

**Interpretasi**<br>
Grafik di atas menunjukkan bahwa 3 kualitas air yang paling berpengaruh terhadap FCR adalah Tranprency, evening_temperature, dan morning_salinity. Artinya FCR yang semakin rendah dapat diperoleh dari Tranprency, evening_temperature, dan morning_salinity yang tinggi.
